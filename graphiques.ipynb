```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
### forme générale ###
d = [10,20,30,40,50]
n40 = [0.55980,0.38945,0.32933,0.32401,0.32397]

plt.plot(d,n40,label="#estimators=40")

plt.title("shape of error evolution")
plt.xlabel("max depth")
plt.ylabel("error")
plt.legend()
plt.show()
```

```python
### TOUS LES TESTS SANS CLUSTERING ###
d = [30,40,50]
n40  = [0.32933,0.32401,0.32397]
n50  = [0.32848,0.32290,0.32283]
n60  = [0.32800,0.32232,0.32222]
n70  = [0.32767,0.32185,0.32176]
n80  = [0.32742,0.32151,0.32138]
n90  = [0.32716,0.32120,0.32106]
n100 = [0.32699,0.32093,0.32078]

plt.plot(d,n40,label="#estimators=40")
plt.plot(d,n50,label="#estimators=50")
plt.plot(d,n60,label="#estimators=60")
plt.plot(d,n70,label="#estimators=70")
plt.plot(d,n80,label="#estimators=80")
plt.plot(d,n90,label="#estimators=90")
plt.plot(d,n100,label="#estimators=100")

plt.title("shape of error evolution for severals number of estimators")
plt.xlabel("max depth")
plt.ylabel("error")
plt.legend()
plt.show()
```

```python
### AJOUT DU CLUSTERING ###
d = [40,50]

withoutClusters = [0.32093,0.32078]
withClusters = [0.32021,0.32010]

plt.plot(d,withoutClusters,label="without clusters")
plt.plot(d,withClusters,label="with clusters")

plt.title("shape of error evolution for best predictors with and without clustering")
plt.xlabel("max depth")
plt.ylabel("error")
plt.legend()
plt.show()
```

```python
size = [1000,5000,8000]
ssCrossError = [0.51638,0.43011,0.41881]
ssValError   = [0.49160,0.43134,0.41532]

plt.plot(size,ssCrossError,label="selection error",c='red')
plt.plot(size,ssValError,label="test error",c='green')

plt.title("no mean, variance = 1")
plt.xlabel("training size")
plt.ylabel("error")
plt.legend()
plt.show()
```
